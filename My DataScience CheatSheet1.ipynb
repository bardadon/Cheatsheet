```python
##The point here is to make as many generic functions as possible in order to reuse them in the future
##this function are not specific to the salary project and can be used in every project(with a few changes)

def load_file(file):
    '''loads csv to pd dataframe'''
    return pd.read_csv(file)

def consolidate_data(df1, df2, key=None, left_index=False, right_index=False):
    '''perform inner join to return only records that are present in both dataframes'''
    return pd.merge(left=df1, right=df2, how='inner', on=key, left_index=left_index, right_index=right_index)

def clean_data(raw_df):
    '''remove rows that contain salary <= 0 or duplicate job IDs'''
    clean_df = raw_df.drop_duplicates(subset='jobId')
    clean_df = clean_df[clean_df.salary>0]
    return clean_df

def one_hot_encode_feature_df(df, cat_vars=None, num_vars=None):
    '''performs one-hot encoding on all categorical variables and combines result with continous variables'''
    cat_df = pd.get_dummies(df[cat_vars])
    num_df = df[num_vars].apply(pd.to_numeric)
    return pd.concat([cat_df, num_df], axis=1)#,ignore_index=False)

def get_target_df(df, target):
    '''returns target dataframe'''
    return df[target]

def train_model(model, feature_df, target_df, num_procs, mean_mse, cv_std):
    neg_mse = cross_val_score(model, feature_df, target_df, cv=2, n_jobs=num_procs, scoring='neg_mean_squared_error')
    mean_mse[model] = -1.0*np.mean(neg_mse)
    cv_std[model] = np.std(neg_mse)

def print_summary(model, mean_mse, cv_std):
    print('\nModel:\n', model)
    print('Average MSE:\n', mean_mse[model])
    print('Standard deviation during CV:\n', cv_std[model])

def save_results(model, mean_mse, predictions, feature_importances):
    '''saves model, model summary, feature importances, and predictions'''
    with open('model.txt', 'w') as file:
        file.write(str(model))
    feature_importances.to_csv('feature_importances.csv') 
    np.savetxt('predictions.csv', predictions, delimiter=',')
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

### Comparing the CDF of a variable to the CDF of a Normal Distribution

- Most of the time, we want features that are as close to a normal distribution as possbile.
- And if they are not distributed normally, we most of the time want to transform them using log, exp, sqrt and so on to make them so.


- Why should we want a feature to be normal distributed?

In Machine Learning, data satisfying Normal Distribution is beneficial for model building. It makes math easier. Models like LDA, Gaussian Naive Bayes, Logistic Regression, Linear Regression, etc., are explicitly calculated from the assumption that the distribution is a bivariate or multivariate normal

https://medium.com/analytics-vidhya/normal-distribution-and-machine-learning-ec9d3ca05070

- We can think about it like this:
- The main goal in Machine Learning is to find the most appropriate distribution that can describe the features we want to describe.
- We find that distribution by building a model and tuning it with some algorithms.
- Features in the real world __usually__ follow the normal distribution because of the Central limit theorem
- That is why a lot of the underlying assumptions of the models we build require normallity.
- So the closest the feature is to normal distribution the better.

- Using a CDF we can check how closly a feature resembels the normal distribution.
- By plotting the normal distribution cdf combined with the cdf of the feature we want to evaluate we can compare between them and see how much does the feature trully resembels the normal distribution

#### In Addition 

One reason the normal distribution is important is that many psychological and educational variables are distributed approximately normally. 

Measures of reading ability, introversion, job satisfaction, and memory are among the many psychological variables approximately normally distributed. 

Although the distributions are only approximately normal, they are usually quite close. A second reason the normal distribution is so important is that it is easy for mathematical statisticians to work with. 

This means that many kinds of statistical tests can be derived for normal distributions. Almost all statistical tests discussed in this text assume normal distributions. 

Fortunately, these tests work very well even if the distribution is only approximately normally distributed. Some tests work well even with very wide deviations from normality. 

Finally, if the mean and standard deviation of a normal distribution are known, it is easy to convert back and forth from raw scores to percentiles.

### We can create a cdf in a different way:

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

#### Bernouli Trial

- lets simulate the situation of 4 coin tosses that each one of them landed a Heads:

```python
np.random.seed(42)
coin_tosses = np.random.random(4)
coin_tosses
```

```python
number_of_heads = 0
for toss in range(len(coin_tosses)):
    if coin_tosses[toss] >= 0.5:
        number_of_heads += 1
        
number_of_heads
```

- So in this particular case, after 4 coin tosses we got 3 heads.
- In order to get the probability to get 4 heads in 4 tosses, lets simulate this experiment a 100000 times and find the proportions of the number of times we got 4 heads:

```python
temp = []
for experiment in range(100000):
    coin_tosses = np.random.random(4)
    number_of_heads = 0
    for coin in range(len(coin_tosses)):
        if coin_tosses[coin] >= 0.5:
            number_of_heads += 1
    if number_of_heads == 4:
        temp.append(True)
    else:
        temp.append(False)
```

```python
sample_mean = sum(temp) / len(temp)
sample_mean
```

```python
population_mean = 0.5**4
population_mean
```

- So the probability to get 4 heads according to our experiment is 0.0627
- And we know that the real proability is 0.0625
- That is pretty close to the actual parameter, and if we were to take more than 100k trials, the probability would be even more accurate

- We can write the Bernouli Trial as a function:

```python
def perform_bernouli_trials(n, p):
    temp = []
    for i in range(n):
        if np.random.random() < p:
            temp.append(True)
        else:
            temp.append(False)
    return sum(temp)
```

```python
perform_bernouli_trials(1000, 0.5)
```

- In other words, out of 1000 bernouli trials where the probability for success is 0.5, 505 experiments ended in success.

- We can use it, to understand the behaviour of certrain situations, for example, in the experiment before that we found that the probability for 4 heads in a row is 0.0627, lets see how many 4 heads in a row we will get from a 10000 trials

```python
perform_bernouli_trials(10000, 0.0627)
```

- 652 out of 10000 trials where each is a 4 coins toss will end in 4 heads

- Correlation is another metric that we can use to describe the relationship between 2 variables
- In statistics, when referring to correlation we usually talk about the Pearson's Correlation Coefficient
- The Persons Coefficient take values from -1 to 1
- We can calculate a correlation matrix by using pd.corr():

```python
age_weight_df = pd.DataFrame({'Age':[10,20,30,40,50,60,70,80,90],
                             'Weight':[30,70,80,80,85,75,65,55,45]})
age_weight_df
```

- Lets check the correlation between Age and Weight:

```python
age_weight_df.corr()
```

- As we can see there is no correlation between Age and Weight.
- It's tempting to say that there's no predictive power between Weight and Age since the correlation is -0.04, but of course, that is wrong, Age can be a pretty good predictor for Weight
- So why is the correlation so low?
- The relationship between Age and Weight is this:

```python
sns.regplot(data = age_weight_df, x = 'Age', y =  'Weight')
```

- The reason the correlation was so low is that correlation is a measure for Linear Relationships, and as it turns out the relationship here is Non-Linear, that is why it got a low correlation score.
- When the relationship is Non-Linear, correlation generally underestimates how strong it is

- Using correlation in this case(Non-Linear Relationship) is useless.
- To Summarize:
    - If the correlation is equal or close to 1 or -1: Strong Linear Relationship
    - if the correlation is equal or close to 0: Weak Linear Relationship, 
    BUT UNKNOWN NON-LINEAR RELATIONSHIP —→ We should further check the relationship before dismissing it predictive power.
- And that is one of the reasons correlation is probably not a great statistics to use.
- Also, correlation doesn't give any information about the slope of the relationship between x and y
- Meaning it doesnt tell us anything about how fast y rises when x rises, in other words, theres no information about how strong the relationship is, only how linear it is.

## <ins>__ETL Process__<ins>

```python
%config Completer.use_jedi = False
import glob
import numpy as np
import pandas as pd
```

### Extract 
- From Json & CSV

```python
def extract_from_csv(file_to_process):
    '''
    Extract Data From a CSV File
    Args:
        CSV File
    Returns:
        Pandas Dataframe
    '''
    
    df = pd.read_csv(file_to_process)
    return df

def extract_from_json(file_to_process):
    '''
    Extract Data From a JSON File
    Args:
        JSON File
    Returns:
        Pandas Dataframe
    '''
    df = pd.read_json(file_to_process,lines=True)
    return df

def extract():
    '''
    Aggregates data from CSV and JSON files to a Dataframe
    Args:
    Returns:
        Pandas Dataframe
    '''
    # Create an Empty DataFrame
    extracted_data = pd.DataFrame() # create an empty data frame to hold extracted data
    
    # Process All CSV Files
    csv_files_paths = glob.glob("*.csv")
    for csv_file in csv_files_paths:
        extracted_data = extracted_data.append(extract_from_csv(csv_file), ignore_index=True)
        
    
    # Process All JSON Files
    json_files_paths = glob.glob("*.json")
    for json_file in json_files_paths:
        extracted_data = extracted_data.append(extract_from_json(json_file), ignore_index=True)
        
    
    return extracted_data
extract()
```

### Transform


```python
def transform(data):
        #Convert height which is in inches to millimeter
        #Convert the datatype of the column into float
        #data.height = data.height.astype(float)
        #Convert inches to meters and round off to two decimals(one inch is 0.0254 meters)
        data['height'] = round(data.height * 0.0254,2)
        
        #Convert weight which is in pounds to kilograms
        #Convert the datatype of the column into float
        #data.weight = data.weight.astype(float)
        #Convert pounds to kilograms and round off to two decimals(one pound is 0.45359237 kilograms)
        data['weight'] = round(data.weight * 0.45359237,2)
        return data

transform(extract())
```

### Load


```python
def load(targetfile, data_to_load):
    '''
    Loading a Pandas Dataframe to a file.
    
    Args:
        1.targetfile: the path of the file.
        2.data_to_load: the data to load into the file.
        
    Returns:
        
    '''
    data_to_load.to_csv(targetfile)

targetfile = 'Transformed_data.csv'
```

```python
load(targetfile,transform(extract()))
```

### Create a Log file


```python
from datetime import datetime

def log(message):
    '''
    Create a timestamp log in the form of '%Y-%h-%d-%H:%M:%S'
    Append the timesmap and a message to a text file called logfile.txt
    
    Args:
        message
    Returns:
        
    '''
    timestamp_format = '%Y-%h-%d-%H:%M:%S'
    now = datetime.now()
    timestamp = now.strftime(timestamp_format)
    with open('logfile.txt', 'a') as f:
        f.write(timestamp + ',' + message + '\n')
```

### Run the ETL Process 


```python
log('Extracting Data')
extract()

log('Transforming Data')
transform(extract())

log('Loading Data')
load(targetfile,transform(extract()))
```

### Cheatsheet

- `docker run <NameOfContainer>` - Run- start a container
    - If the image exists locally it will use it, if not, docker will go-to hub.docker and pull the image from there.
    - That is only for the first time, the next time, the same image will be reused.
- `docker ps` - list all **running** containers
- `docker ps -a` - List all containers
- `docker stop <NameOfContainer>` - Stop a container
- `docker rm <NameOfContainer>`  - Remove a container
- `docker images` - list images
- `docker rmi <NameOfImage>` - Remove an image
    - We must stop and delete all dependent containers to be able to delete an image.
- `docker pull <NameOfImage>`  - Pull and image(without running the container)
    - The docker run command, will pull and run the container
- `exec` - Execute a command
- `docker exec <NameOfContainer> cat / etc/hosts` - example: output the files under the container at the address etc/hosts
- `docker run kodekloud/simple-webapp`  - Run a container - attach and detach
- `docker run -d kodekloud/simple-webapp`  - Run a container - attach and detach(run it in the background)
    - The output will be something like this: 1ccb35b3aafab1633822a4e4198c69cf09e58567abc73532da4ef11a73d3cceb
- `docker attach 1ccb` - Attach a container

**Stop and remove all images and containers**

1. `docker kill $(docker ps -q)`
2. `docker rm $(docker ps -a -q)`
3. `docker rmi $(docker images -q)`

- `docker run <NameOfContainer>:<Version>` - Run a container with a tag.
    - For example: `docker run redis:4.0`
    - Docker will run a specific version of redis, provided it appears in Hub.docker.
    - Not specifying a tag, means that docker will pull the latest version(the version that is tagged “latest”).
    - To see all the tags, go to Hub.docker and inside every image there will be a “Tags” tab that stores all the different versions of that image.
- `docker run -i <NameOfContainer>` - Run container in an interactive mode(accept inputs from the user)
- `docker run -it <NameOfContainer>` - Run container in an interactive mode(accept inputs from the user)
- `docker run -p <WebServerPort>:<LocalPort> <NameOfContainer>` - Run the container, and map the web server port to a free local port.
    - Example: `docker run -p 80:5000 kodekloud/simple-webapp`
    - Note: We can run the application in a different port simultanisly
        - `docker run -p 8000:5000 kodekloud/simple-webapp`
    - Note: port 80 is **the port that the server "listens to" or expects to receive from a Web client**
    - Note: port 5000 is used to **to accept incoming connections from other UPnP devices**
- `docker run -v <DockerHostLocation>:<DockerContainerLocation> <NameOfContainer>`- Map the location of the data in the container, to a directory in the host. If the database container will be deleted, the data will move to the new location outside of the container.
    - Example: `docker run -v /opt/datadir:/var/lib/mysql mysql`
    - Now when I delete the mysql container, data will be moved from: `/var/lib/mysql` to `/opt/datadir`.
- `docker inspect <NameOfContainer>` - Inspect and provide details about a container.
- `docker logs <NameOfContainer>` -   See logs of running containers
- `docker run -e <NameOfEnvVariable>; <NameOfContainer>` - Run the container and specify the value of an environment variable.

# <ins>__Airflow__<ins>

## Step 1: Importing modules

```python
from datetime import timedelta

import airflow
from airflow import DAG
from airflow.operators.bash_operator import BashOperator
```

## Step 2: Default Arguments

```python
# Define default and DAG-specific arguments
default_args = {
    'owner': 'airflow',    
    'start_date': airflow.utils.dates.days_ago(2),
    # 'end_date': datetime(2018, 12, 30),
    'depends_on_past': False,
    'email': ['airflow@example.com'],
    'email_on_failure': False,
    'email_on_retry': False,
    # If a task fails, retry it once after waiting
    # at least 5 minutes
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
    }
```

## Step 3: Instantiate a DAG

```python
# Give the DAG name, configure the schedule, and set the DAG settings
dag = DAG(
    'tutorial',
    default_args=default_args,
    description='A simple tutorial DAG',
    # Continue to run DAG once per day
    schedule_interval=timedelta(days=1),
)
```

- Here is a couple of options you can use for your schedule_interval. You can choose to use some preset argument or cron-like argument:

```python
from IPython.display import Image
Image(filename='image.png') 
```

## Step 4: Tasks

```python
# The next step is to lay out all the tasks in the workflow.
# t1, t2 and t3 are examples of tasks created by instantiating operators
t1 = BashOperator(
    task_id='print_date',
    bash_command='date',
    dag=dag,
)

t2 = BashOperator(
    task_id='sleep',
    depends_on_past=False,
    bash_command='sleep 5',
    dag=dag,
)

templated_command = """
{% for i in range(5) %}
    echo "{{ ds }}"
    echo "{{ macros.ds_add(ds, 7)}}"
    echo "{{ params.my_param }}"
{% endfor %}
"""

t3 = BashOperator(
    task_id='templated',
    depends_on_past=False,
    bash_command=templated_command,
    params={'my_param': 'Parameter I passed in'},
    dag=dag,
)
```

## Step 5: Setting up Dependencies


- Set the dependencies or the order in which the tasks should be executed.
- Here are a few ways you can define dependencies between them:


```python
# This means that t2 will depend on t1
# running successfully to run.
t1.set_downstream(t2)

# similar to above where t3 will depend on t1
t3.set_upstream(t1)
```

```python
# The bit shift operator can also be
# used to chain operations:
t1 >> t2

# And the upstream dependency with the
# bit shift operator:
t2 << t1
```

```python
# A list of tasks can also be set as
# dependencies. These operations
# all have the same effect:
t1.set_downstream([t2, t3])
t1 >> [t2, t3]
[t2, t3] << t1
```

### Recap

- Basically, a DAG is just a Python file, which is used to organize tasks and set their execution context. DAGs do not perform any actual computation.
- Instead, tasks are the element of Airflow that actually “do the work” we want to be performed. And it is your job to write the configuration and organize the tasks in specific orders to create a complete data pipeline.
- Your final DAG will look like this:


```python
"""
Code that goes along with the Airflow tutorial located at:
https://github.com/apache/incubator-airflow/blob/master/airflow/example_dags/tutorial.py
"""
from airflow import DAG
from airflow.operators.bash_operator import BashOperator
from datetime import datetime, timedelta


default_args = {
    'owner': 'airflow',
    'depends_on_past': False,
    'start_date': datetime(2015, 6, 1),
    'email': ['airflow@example.com'],
    'email_on_failure': False,
    'email_on_retry': False,
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
    # 'queue': 'bash_queue',
    # 'pool': 'backfill',
    # 'priority_weight': 10,
    # 'end_date': datetime(2016, 1, 1),
}

dag = DAG(
    'tutorial', 
    default_args=default_args, 
    schedule_interval=timedelta(days=1))

# t1, t2 and t3 are examples of tasks 
# created by instantiating operators
t1 = BashOperator(
    task_id='print_date',
    bash_command='date',
    dag=dag)

t2 = BashOperator(
    task_id='sleep',
    bash_command='sleep 5',
    retries=3,
    dag=dag)

templated_command = """
    {% for i in range(5) %}
        echo "{{ ds }}"
        echo "{{ macros.ds_add(ds, 7)}}"
        echo "{{ params.my_param }}"
    {% endfor %}
"""

t3 = BashOperator(
    task_id='templated',
    bash_command=templated_command,
    params={'my_param': 'Parameter I passed in'},
    dag=dag)

t2.set_upstream(t1)
t3.set_upstream(t1)
```
