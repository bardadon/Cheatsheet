```python
# Set the pattern to search to be 'abc'
pattern = re.compile(r'\.')

# Find all matches
matches = pattern.finditer(text_to_search)

# Create a Match object for the pattern
for match in matches:
    print(match)
```

## Python Tutorial: Generators - How to use them and the benefits you receive


- Generators are iterables in Python that do not hold the data in memory.
- This allow us to run programs more efficiently.
- Bascially, we hold values in a generator, and only when we need them, we extact them from the generator(and store them in the memroy)

```python
# Normal Function
def square_numbers(nums):
    result = []
    
    for i in nums:
        result.append(i*i)
    return result
```

```python
my_nums = square_numbers([1,2,3,4])
my_nums
```

```python
# Generator
def square_numbers(nums):    
    for i in nums:
        yield i*i
```

```python
my_nums = square_numbers([1,2,3,4])
my_nums
```

- As we can see, my_nums is a generator object. 
- Generators dont hold the entire result at once.

## Python Tutorial: Decorators - Dynamically Alter The Functionality Of Your Functions


### Recap of first-class functions and closures

### Creating a decorator

- A decorator is a function that takes another function as an argument.
- It add somekind of functionality to that function 
- And returns a function
- All of this without altering the source code of function 

```python
def decorator_function(original_function):
    
    def wrapper_function():
        return original_function()
        
    return wrapper_function
```

- In this example i have created a function called decorator_function that takes a function as an argument and returns that function.
- Now we can pass functions into that function, for example:

```python
def display():
    print('display function ran')
    
decorated_display = decorator_function(display)
```

```python
decorated_display()
```

- This was the basic template for wrapping a function in another function.
- Lets actually add some functionality to the functions that i insert to decorator_function()

```python
def decorator_function(original_function):
    
    def wrapper_function():
        print('wrapper executed this before {}'.format(original_function.__name__))
        return original_function()
        
    return wrapper_function
```

```python
decorated_display = decorator_function(display)
decorated_display()
```

- As we can see, now the function has an added functionality.
- We can get the same result by using decorator_function as an actual decorator instead of passing functions to it as arguments

```python
@decorator_function
def display():
    print('display function ran')
```

- Now display will automatically have that added functionality:

```python
decorated_display = display()
decorated_display
```

### Adding *args and **kwargs

- Lets try to add this decorator to other functions, one that takes an argument:

- This is the result of the function:

```python
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- Now lets add the decorator:

- As we can see we get an error.
- We want to be able to pass any arbitrary amount of arguments to the wrapper function.
- We can do that by adding *args and **kwargs as arguments:

```python
def decorator_function(original_function):
    
    def wrapper_function(*args, **kwargs):
        print('wrapper executed this before {}'.format(original_function.__name__))
        return original_function(*args, **kwargs)
        
    return wrapper_function
```

```python
@decorator_function
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- Now this works

### Class decorator

- We can do the same with a decorator class instead of a decorator function

```python
class decorator_class():
    
    def __init__(self, original_function):
        self.original_function = original_function
        
    def __call__(self, *args, **kwargs):
        print('wrapper executed this before {}'.format(self.original_function.__name__))
        return self.original_function(*args, **kwargs)
```

```python
@decorator_class
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- As we can see, we get the same result but now we are decorating with a class instead of a function

### Logging and timing with decorators

- One of the more common use case for decorators is to create logs.
- Ill use the library logging to create a decorator that creates .log files  
- For example:

```python
def my_logger(orig_func):
    
    # Adding functionality
    import logging # Import logging module
    logging.basicConfig(filename='{}.log'.format(orig_func.__name__), level = logging.INFO) # Create a .log file
    
    # Creating the decorator
    def wrapper(*args, **kwargs):
        
        # Printing a sentence to the .log file
        logging.info( 
        'Ran with args: {}, and kwargs: {}'.format(args, kwargs))
        return orig_func(*args, **kwargs)
    
    return wrapper
```

```python
@my_logger
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- Now the function also created a .log file with the content:

INFO:root:Ran with args: ('Bar', 30), and kwargs: {}


- This is a great example of why decorators can be useful.
- Instead of having to manually wrap  each function that we want to log in a code to log it, we can simply decorate it with @my_logger

- Another very common use case is to time the function with a decorator

```python
def my_timer(orig_func):
    
    # Import time
    import time
    
    def wrapper(*args, **kwargs):
        
        # Set current time
        t1 = time.time()
        
        # Run the function
        result = orig_func(*args, **kwargs)
        
        # Calculate time difference and print statement
        t2 = time.time() - t1
        print('{} ran in: {} sec'.format(orig_func.__name__, t2))
        
        return result
    
    return wrapper
```

```python
@my_timer
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

```python
@my_timer
def display_info(name, age):
    import time
    time.sleep(1)
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- We can also use both the logger and the timer:

```python
@my_logger
@my_timer
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- The order of the decorators matter.
- This

```python
@my_timer
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

- is equal to this

```python
display_info = my_timer(display_info)
```

- is equal to this

```python
display_info = my_logger(my_timer(display_info))
```

```python
@my_timer
@my_logger
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

- Its a best practice to add the decorator @wraps(orig_func) to the wrapper function

```python
from functools import wraps
```

```python
def my_logger(orig_func):
    
    # Adding functionality
    import logging # Import logging module
    logging.basicConfig(filename='{}.log'.format(orig_func.__name__), level = logging.INFO) # Create a .log file
    
    # Creating the decorator
    @wraps(orig_func)
    def wrapper(*args, **kwargs):
        
        # Printing a sentence to the .log file
        logging.info( 
        'Ran with args: {}, and kwargs: {}'.format(args, kwargs))
        return orig_func(*args, **kwargs)
    
    return wrapper
```

```python
def my_timer(orig_func):
    
    # Import time
    import time
    
    @wraps(orig_func)
    def wrapper(*args, **kwargs):
        
        # Set current time
        t1 = time.time()
        
        # Run the function
        result = orig_func(*args, **kwargs)
        
        # Calculate time difference and print statement
        t2 = time.time() - t1
        print('{} ran in: {} sec'.format(orig_func.__name__, t2))
        
        return result
    
    return wrapper
```

```python
@my_timer
@my_logger
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info.__name__
```

- As we can see, the name of the function now is the corrent name: display_info

- using the functool.wraps let us retain the information of the original function, its name, docstring and so on.

### Decorators with arguments

- Ill create a simple decorator that prints: Executed Before with the function name, run the function, and Executed After with the function name.
- Ill also add *args and **kwargs to let the wrapper accept any number of arguments

```python
# Decorators
def decorator_function(original_function):
    def wrapper_function(*args, **kwargs):
        print('Executed Before', original_function.__name__)
        result = original_function(*args, **kwargs)
        print('Executed After', original_function.__name__, '\n')
        return result
    return wrapper_function

```

```python
@decorator_function
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

- Now lets add some arguments to the decorator.
- Ill wrap the entire decorator_function with a function called prefix_decorator.
- This will allow me to add an extra argument to add to the funtionality.
- This argument can be added to the decorator:

```python
def prefix_decorator(prefix):
    
    def decorator_function(original_function):
        def wrapper_function(*args, **kwargs):
            print(prefix, 'Executed Before', original_function.__name__)
            result = original_function(*args, **kwargs)
            print(prefix, 'Executed After', original_function.__name__, '\n')
            return result
        return wrapper_function
    
    return decorator_function
```

```python
@prefix_decorator('LOG:')
def display_info(name, age):
    print('display_info ran with arguments ({}, {})'.format(name, age))
```

```python
display_info('Bar', 30)
```

## Python Tutorial: Namedtuple - When and why should you use namedtuples?


- A Namedtuple is a lighter and more readable version of a regular tuple
- Example, lets represent RGB in a tuple:

```python
color = (55,155,255)
```

```python
color[0]
```

- The issue with this is that the numbers dont have any labels to them. So if ill come to this code in the future and see the output 55, i wont be able to understand what these numbers represent.
- Instead we could represent it with a dictionary:

```python
color = {'Red': 55,
         'Green': 155,
         'Blue': 255}
```

```python
color[0]
```

- As we can see, we get the same output, only now the items have keys to label them
- The problem with dictionaries is that they require a lot of typing, and is not very flexible to changes
- Lets see why using namedTuple is useful:

```python
from collections import namedtuple
color = namedtuple('color', ["Red",'Green','Blue'])
random_color = color(55,155,255)
```

```python
random_color
```

- As we can see, this is a lot more readable.
- We can access the values the same as dictionaries

```python
random_color.Red
```

```python
random_color[0]
```

- Also this is more flexible, We can easily create a color from this named tuple, lets create the color white:

```python
white = color(255,255,255)
```

```python
white
```

```python
print('The blue value of White is: {}'.format(white.Blue))
```

- With a dictionary, if we want to creat the color white, we would need to manually create a whole new dictionary:

```python
white = {'Red': 255,
         'Green': 255,
         'Blue': 255}

white
```

- If the response from the Linode server where the app is running is not 200:
    - Run the function notify_user() which sends an email to me.
    - Run the function reboot_server() which loads a new instance of the server.


```python
import os
import smtplib
import requests
# import logging
from linode_api4 import LinodeClient, Instance

EMAIL_ADDRESS = os.environ.get('EMAIL_USER')
EMAIL_PASSWORD = os.environ.get('EMAIL_PASS')
LINODE_TOKEN = os.environ.get('LINODE_TOKEN')

# logging.basicConfig(filename='PATH_TO_DESIRED_LOG_FILE',
#                     level=logging.INFO,
#                     format='%(asctime)s:%(levelname)s:%(message)s')


def notify_user():
    with smtplib.SMTP('smtp.gmail.com', 587) as smtp:
        smtp.ehlo()
        smtp.starttls()
        smtp.ehlo()

        smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)

        subject = 'YOUR SITE IS DOWN!'
        body = 'Make sure the server restarted and it is back up'
        msg = f'Subject: {subject}\n\n{body}'

        # logging.info('Sending Email...')
        smtp.sendmail(EMAIL_ADDRESS, 'INSERT_RECEIVER_ADDRESS', msg)


def reboot_server():
    client = LinodeClient(LINODE_TOKEN)
    my_server = client.load(Instance, 376715)
    my_server.reboot()
    # logging.info('Attempting to reboot server...')


try:
    r = requests.get('https://example.com', timeout=5)

    if r.status_code != 200:
        # logging.info('Website is DOWN!')
        notify_user()
        reboot_server()
    else:
        # logging.info('Website is UP')
except Exception as e:
    # logging.info('Website is DOWN!')
    notify_user()
    reboot_server()
```

### Improving the interface

- Ill create a few function that will assit with the handling of the database.
- Ill also pass the c.execure part as a context manager, that will auto commit the transactions and we will not have to remember to close the connections for now on.

```python
# INSERT
def insert_emp(emp):
    with conn:
        c.execute('insert into employees values (:first, :last, :pay)',{'first': emp.first, 'last':emp.last, 'pay':emp.pay})
        

# SELECT
def get_emps_by_name(lastname):
    c.execute("select * from employees where lastname = :last", {'last': lastname})
    return c.fetchall()

# UPDATE
def update_pay(emp, pay):
    with conn:
        c.execute("""UPDATE employees SET pay = :pay
                    WHERE firstname = :first AND lastname = :last""",
                  {'first': emp.first, 'last': emp.last, 'pay': pay})

# DELETE
def remove_emp(emp):
    with conn:
        c.execute("DELETE from employees WHERE firstname = :first AND lastname = :last",
                  {'first': emp.first, 'last': emp.last})
```

## Python Tutorial: Logging Basics - Logging to Files, Setting Levels, and Formatting


- Logging is an extremely useful tool when writing programs. Instead of relying on prints throughout the script to inform us and help us troubleshoot we should use logging instead.
- There are 5 typical modes to a log:

##### DEBUG: Detailed information, typically of interest only when diagnosing problems.

##### INFO: Confirmation that things are working as expected.

##### WARNING: An indication that something unexpected happened, or indicative of some problem in the near future (e.g. disk space low). The software is still working as expected.

##### ERROR: Due to a more serious problem, the software has not been able to perform some function.

##### CRITICAL: A serious error, indicating that the program itself may be unable to continue running.

- Depending on the use-case we should choose the level of details in the log according to those 5 modes.
- The default level is WARNING. Meaning, only when something unexpected happened should we insert a log.
- Some use cases would require us the log every single action for future troubleshooting.

### Creating a logger

- Earlier I had issues where i would run a script with a new log file, but the content of the logs got appended to old log file.
- That happended because logs are appended to the root log file(on some conditions)
- Its a best practice to specify where and which logger we want to be relevant for the script by creating the object: logger.
- Then specify the formatter(the form of the logs) and handler(the log file) like shown below.
- This will replace the basic config from before.
- Also, we use that logger.info instead of logging.info where ever we want logs.

```python
import logging


logger = logging.getLogger(__name__)

formatter = logging.Formatter('%(asctime)s:%(levelname)s:%(message)s')

file_handler = logging.FileHandler('employee1.log')

file_handler.setFormatter(formatter)

logger.addHandler(file_handler)

class Employee:
    """A sample Employee class"""

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        
        logger.info('Created Employee: {} - {} - {}'.format(self.first, self.last, self.pay))


    @property
    def email(self):
        return '{}.{}@email.com'.format(self.first, self.last)

    @property
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
```

## Python Tutorial: Else Clauses on Loops


```python
my_list = [1,2,3,4,5]

for i in my_list:
    print(i)
else:
    print('Hit the For/Else Statement')
```

- As we can see, running this script is actually running both the print(i) and the print('Hit the For/Else Statement').
- I.E both the if and the else got printed.
- We should actually think about else in for loops as "not break".
- It functions as the opposite of break
- So to transalte this script:


"print i for every item in my_list, then, if you are able to finish everything, print 'Hit the For/Else Statement'"

- For example:

```python
my_list = [1,2,3,4,5]

for i in my_list:
    print(i)
    if i == 6:
        break
else:
    print("Able to finish everything, printing the statement")
```

```python
my_list = [1,2,3,4,5]

for i in my_list:
    print(i)
    if i == 3:
        break
else:
    print("Unable to finish everything, NOT printing the statement")
```

### Creating our own iterator using classes

- Lets try to create our own iterator:

- Notes:
    - Every iterator is also an iterable
    - Not every iterable is also an iterator.
    - For example a list is an iterable and not an iterator.
    - To create an iterable we need to add the method __iter __
    - To futhur it and make the iterable also an iterator, we need to add the method __next __

```python
class MyRange:
    
    def __init__(self, start, end):
        self.start = start
        self.end = end
    
    def __iter__(self):
        return self
    
    def __next__(self):
        
        if self.start >= self.end :
            raise StopIteration
            
        current = self.start
        self.start += 1
        return current
```

```python
nums = MyRange(start=1, end=4)
```

```python
for num in nums:
    print(num)
```

```python
import time

start = time.perf_counter()

def do_something(seconds):
    print(f'Sleeping {seconds} second(s)...')
    time.sleep(seconds)
    return f'Done Sleeping...{seconds}'

do_something(1)
do_something(1)

finish = time.perf_counter()

print(f'Finished in {round(finish-start, 2)} second(s)')
```

```python
Image.open('sequential coding.PNG')
```

- Now, during these sleeping periods, the CPU doesnt do anything, it just waiting.
- In general, coding tasks divides into 2, CPU bound tasks and IO bound tasks.

- CPU bound tasks are tasks that require heavy calculations from the CPU, during those tasks the CPU is overloaded and is working at full capcacity.
- IO bound tasks are tasks that require/waiting for an input before producing an output, they are waiting for some other task to be completed and are not using the CPU all that much
- IO Example :reading and writing from a disk, download stuff online

- If the our task is CPU bound, we are not going to get that much benefit from making the CPU perform parallel calculcations, it is already at full capcacity
- In some instances it will actually hurt the performance to thread such a heavy CPU bound task because of the extra overhead that threading can cause.

- On the other hand, if a task is IO bound, we will actually get much benefit from threading the task and calculating it in parallel

- The whole idea of threading is not actually to perform calculation in parallel, its just looks like that, its more like performing calculations when other tasks are waiting for an input.
- Lets picture the task from above, only this time in "parallel":

```python
import time
import concurrent.futures
from PIL import Image, ImageFilter

img_names = [
    'photo-1516117172878-fd2c41f4a759.jpg',
    'photo-1532009324734-20a7a5813719.jpg',
    'photo-1524429656589-6633a470097c.jpg',
    'photo-1530224264768-7ff8c1789d79.jpg',
    'photo-1564135624576-c5c88640f235.jpg',
    'photo-1541698444083-023c97d3f4b6.jpg',
    'photo-1522364723953-452d3431c267.jpg',
    'photo-1513938709626-033611b8cc03.jpg',
    'photo-1507143550189-fed454f93097.jpg',
    'photo-1493976040374-85c8e12f0c0e.jpg',
    'photo-1504198453319-5ce911bafcde.jpg',
    'photo-1530122037265-a5f1f91d3b99.jpg',
    'photo-1516972810927-80185027ca84.jpg',
    'photo-1550439062-609e1531270e.jpg',
    'photo-1549692520-acc6669e2f0c.jpg'
]

t1 = time.perf_counter()

size = (1200, 1200)


def process_image(img_name):
    img = Image.open(img_name)

    img = img.filter(ImageFilter.GaussianBlur(15))

    img.thumbnail(size)
    img.save(f'processed/{img_name}')
    print(f'{img_name} was processed...')


with concurrent.futures.ProcessPoolExecutor() as executor:
    executor.map(process_image, img_names)


t2 = time.perf_counter()

print(f'Finished in {t2-t1} seconds')
```

    - If a task is cpu-bound, multi-process it by using the context manager: concurrent.futures.ProcessPoolExecutor()

### Class Methods

- Another type of methods is when we pass the class as the first argument

- We can make a regular methods into a class method by adding the decorator: @classmethod, for example, lets make another apply_raise method, but this time its going to be a class method instead of a regular one.

```python
class Employee():
    
    num_of_employee = 0
    percent_raise = 1.04

    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
        Employee.num_of_employee += 1
    
    def print_full_name(self):
        print(self.first + ' ' + self.last)
        
    
    def apply_raise(self):
        self.pay = self.pay * self.percent_raise
        
    @classmethod    
    def set_raise_amount(cls,amount):
        cls.percent_raise = amount
```

- Just like "self" is the common convension name for the actual instance of the Class(i.e emp_1)
- So does, "cls" is the common convension name for the argument class. we cant use Class because that is a saved word, so instead we will use cls.

### Static Methods

- If we do not intend to use class or self variables. we can create a static method.
- Static do not take self or class as arguments, they are basically regualr functions inside of a class, the only reason that we put them in the class and not outside, is because they have some logical connection to the site.

- Lets create a static method, that takes a date and return wether or not its a work day.
- This function/method, as we can see, is not related to a specifc instance(i.e a specifc employee) or does it belong to the entire class.
- However, it is relevant for the use case of the class, so we should put it inside the class as a static method, as such:

```python
class Employee():
    
    from datetime import datetime
    
    num_of_employee = 0
    percent_raise = 1.04

    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
        Employee.num_of_employee += 1
    
    def print_full_name(self):
        print(self.first + ' ' + self.last)
        
    
    def apply_raise(self):
        self.pay = int(self.pay * self.percent_raise)
        
    @classmethod    
    def set_raise_amount(cls,amount):
        cls.percent_raise = amount
        
        
    @classmethod
    def from_string(cls, emp_str):
        first, last, pay = emp_str.split('-')
        return Employee(first, last, pay)
    
    @staticmethod
    def is_workday():
        if datetime.weekday(datetime.today()) != 5 or datetime.weekday(datetime.today()) != 4:
            return True
        else:
            return False
```

```python
emp_str_1 = 'John-Doe-70000'
emp_1 = Employee.from_string(emp_str=emp_str_1)
```

```python
emp_1.is_workday()
```

- One clue that a method should be a static method, is that we dont need to access the class or the instance throughout the method.

```python
class Manager(Employee):
    
    percent_raise = 1.25
    
    def __init__(self, first, last, pay, employees=None):
        super().__init__(first, last ,pay)
        if employees is None:
            self.employees = []
        else:
            self.employees = employees
            
    def add_employee(self, emp):
        self.emp = emp
        
        if emp not in self.employees:
            self.employees.append(emp)
    
    def remove_emp(self, emp):
        self.emp = emp
        
        if emp in self.employees:
            self.employees.remove(emp)
            
    def print_emps(self):
        for emp in self.employees:
            
            print('--->', emp.print_full_name())
```
