```python
left, right = 0, len(nums) - 1

while left <= right:
    mid = (left + right) // 2
    
    if nums[mid] == target:
        print(nums[mid])
        break;
        
    elif nums[mid] < target:
        left = mid + 1
    
    else:
        right = mid - 1
```

```python
nums = [1,5,4,3,8,5,7,53,2,1,57,98]
target = 2

nums = sorted(nums)

def binarySearch(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: int
    """
    if len(nums) == 0:
        return -1

    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # End Condition: left > right
    return -1
```

```python
binarySearch(nums, target)
```

## Insertion Sort

```python
def insertion_sort(nums):
    
    for j in range(2,len(nums)):
        
        key = nums[j]
        i = j - 1
        
        while i > 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i = i - 1
            
        nums[i + 1] = key
        
    return nums
```

```python
nums = [1,5,4,3,8,5,7,53,2,1,57,98]
insertion_sort(nums)
```

```python
class Person():
    
    def __init__(self, firstname, lastname, age, gender):
        self.firstname = firstname
        self.lastname = firstname
        self.age = age
        self.gender = gender
    
    @property
    def new_age():
        return self.age
    
    @new_age.setter
    def new_age(self, new_age):
        self.age = new_age
```

- By wrapping an object with the method delay, we can turn it into a lazy evaluator.
- WE can wrapp multiple functions with delay, and let DASK take care about the optimization of how to run them

```python
from dask import delayed
import numpy as np

def sqrt1(nums1):
    return np.sqrt(nums1)
```

- Without dask, this function will run as soon as i run it

```python
result = sqrt1(4)
result
```

- With Dask, this is now a lazy evaluator and will only be run optimally by Dask when i choose to compute it:

```python
result = delayed(sqrt1(4))
result
```

```python
result.compute()
```

- We can also turn functions into lazy eval using this:

```python
@delayed
def sqrt1(nums1):
    return np.sqrt(nums1)

result = sqrt1(4)
result
```

```python
result.compute()
```

## Majority Element

```python
def majorityElement(nums):
    
    count_dict = dict()
    
    # Count appearances in a dictionary
    for i in nums:

        count_dict[i] = count_dict.get(i,0) + 1
        
    # Grab the key that belongs to the max value
    for key, value in count_dict.items():

        if value == max(count_dict.values()):
            return key
```

```python
from pyspark.sql import SparkSession
import time

def consecutiveNumbers():
    
    t1 = time.perf_counter()
    spark = SparkSession.builder.appName('ConsecutiveNumber').getOrCreate()

    df = spark.read.option('header', 'true').csv('logs.csv')
    df.createOrReplaceTempView('logs')

    query = '''
    select 
        distinct l1.num as ConsecutiveNumber
    from logs as l1
    join logs as l2
    on l1.id = l2.id - 1
    join logs as l3
    on l2.id = l3.id -1
    where 
        l1.num = l2.num and l2.num = l3.num
    '''
    result = spark.sql(query)
    
    t2 = time.perf_counter()
    total_time = t2 - t1
    
    return total_time
```

## Playing with digits

```python
def dig_pow(n, p):

    # Create a generator of digits
    digits_generator = (int(digit) for digit in str(n)) 

    
    digits_sum = 0
    # Calculate the sum of digits with power p
    for digit in digits_generator:
        digits_sum += digit ** p
        p += 1   


    k = 0
    # Raise k by 1, and check if there's a k*n that is equal to digits_sum
    # if there is a k, return it
    while k * n <= digits_sum:
        
        if k*n == digits_sum:
            return k

        k += 1  

    return -1

dig_pow(46288,3)
```
